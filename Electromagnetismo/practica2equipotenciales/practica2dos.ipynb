```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
import matplotlib.colors as mcolors
from mpl_toolkits.axes_grid1.axes_divider import make_axes_locatable
```

```python
predataArray = pd.read_excel("practica2.xlsx", sheet_name="pyin2", header = None, decimal = ",").to_numpy()
```

```python
def insert_row_averages(arr):
    arr = np.asarray(arr)

    # Averages between consecutive rows
    mid = (arr[:-1] + arr[1:]) / 2

    # Prepare output array
    out = np.empty((arr.shape[0] * 2 - 1, arr.shape[1]), dtype=arr.dtype)

    # Fill rows
    out[0::2] = arr
    out[1::2] = mid

    return out
```

```python
dataArray = insert_row_averages(predataArray)
dataArray[3,0:10] = 0
```

```python
largeDataArray = np.hstack((np.fliplr(dataArray[:,1:]), dataArray))
largerDataArray = np.vstack((largeDataArray, -np.flipud(largeDataArray[:-1,:]) + 9.99*2))
largerDataArrayFlipped = np.flipud(largerDataArray)
```

```python
plt.imshow(largerDataArrayFlipped, aspect='equal', cmap = "plasma", origin = "lower", interpolation = "nearest", extent = (0,28,0,20))

plt.show()
```

```python
xGrid = np.arange(0, largerDataArrayFlipped.shape[1], step = 1)
yGrid = np.arange(0, largerDataArrayFlipped.shape[0], step = 1)
```

```python
interp = sp.interpolate.RegularGridInterpolator((yGrid, xGrid), largerDataArrayFlipped, method = "linear")
```

```python
xplot = np.linspace(xGrid.min(), xGrid.max(),560)#*2, 560)
yplot = np.linspace(yGrid.min(), yGrid.max(), 400) #used to be 400

X, Y = np.meshgrid(xplot, yplot, indexing='xy')
data = interp((Y,X))#/2, Y))
```

```python
realSpacingInXplotSpace = xGrid.size/xplot.size
realSpacingInYplotSpace = yGrid.size/yplot.size

fieldGradient = -np.array(np.gradient(data, realSpacingInYplotSpace, realSpacingInXplotSpace)) #unidades de voltio / cm
```

```python
normMatrix = np.sqrt(fieldGradient[0]**2 + fieldGradient[1]**2)
normalizedFieldGradientU = fieldGradient[0]/normMatrix
normalizedFieldGradientV = fieldGradient[1]/normMatrix
```

```python
colors = ["#03071e","#370617","#6a040f","#9d0208","#d00000","#dc2f02","#e85d04","#f48c06","#faa307"]
red_orange_cmap = mpl.colors.LinearSegmentedColormap.from_list("red_orange", colors)
```

```python
fig, ax = plt.subplots(dpi = 250)

divider = make_axes_locatable(ax)


levels = np.linspace(0,20,21)

norm = mcolors.Normalize(vmin = -0, vmax = 20)

img = ax.imshow(data, aspect='equal', cmap = "plasma", origin = "lower", interpolation = "nearest", extent = (0,28,0,20), norm = norm)



CS = ax.contour(X, Y, data, levels = levels, cmap = red_orange_cmap, zorder = 2, norm = norm)
ax.clabel(CS, levels[1:-1], fontsize = 10)

ax.hlines(17, 5, 23 , "gainsboro", lw = 4)
ax.hlines(3, 5, 23, "#232330", lw = 4)



cax = divider.append_axes("right", size="3%", pad=0.2)
cbar = fig.colorbar(img, cax=cax, norm = norm)
cbar.set_label('Voltaje (V)',rotation=-90,  labelpad=15)



ax.set_xticks(np.arange(0, 29, 2))
ax.set_xticks(np.arange(1, 29, 2), minor=True)

ax.tick_params(axis='x', which='major', length=8)
ax.tick_params(axis='x', which='minor', length=4)

ax.set_yticks(np.arange(0, 21, 2))
ax.set_yticks(np.arange(1, 21, 2), minor=True)

ax.tick_params(axis='y', which='major', length=8)
ax.tick_params(axis='y', which='minor', length=4)

ax.grid(color = "#FFFFFF", zorder = 1, alpha = 0.2, ls = (0,(6,7)))

ax.set_xlabel("x (cm)")
ax.set_ylabel("y (cm)")


plt.savefig("casoDosEquipotencialesCondensador.png", dpi = 250, bbox_inches='tight', pad_inches=0.03)
plt.show()
```
