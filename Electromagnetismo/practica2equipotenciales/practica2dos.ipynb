```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
```

```python
predataArray = pd.read_excel("practica2.xlsx", sheet_name="pyin2", header = None, decimal = ",").to_numpy()

#dataArray[1,0] = 0.01
```

```python
def insert_row_averages(arr):
    arr = np.asarray(arr)

    # Averages between consecutive rows
    mid = (arr[:-1] + arr[1:]) / 2

    # Prepare output array
    out = np.empty((arr.shape[0] * 2 - 1, arr.shape[1]), dtype=arr.dtype)

    # Fill rows
    out[0::2] = arr
    out[1::2] = mid

    return out
```

```python
dataArray = insert_row_averages(predataArray)
dataArray[3,0:10] = 0
```

```python
largeDataArray = np.hstack((np.fliplr(dataArray[:,1:]), dataArray))
largerDataArray = np.vstack((largeDataArray, -np.flipud(largeDataArray[:-1,:]) + 9.99*2))
```

```python
xGrid = np.arange(0, largerDataArray.shape[0], step = 1)
yGrid = np.arange(0, largerDataArray.shape[1], step = 1)
```

```python
interp = sp.interpolate.RegularGridInterpolator((xGrid, yGrid), largerDataArray)
```

```python
xplot = np.linspace(xGrid.min(), xGrid.max(),560)#*2, 560)
yplot = np.linspace(yGrid.min(), yGrid.max(), 400)

X, Y = np.meshgrid(xplot, yplot, indexing='ij')
data = interp((X,Y))#/2, Y))
```

```python
fieldGradient = np.gradient(data)
```

```python
normMatrix = np.sqrt(fieldGradient[0]**2 + fieldGradient[1]**2)
normalizedFieldGradientU = fieldGradient[0]/normMatrix
normalizedFieldGradientV = fieldGradient[1]/normMatrix
```

```python
fig, ax = plt.subplots()

ax.imshow(data, aspect='equal', cmap = "cividis", origin = "lower", interpolation = "nearest", extent = (0,28,0,20))
ax.contour(Y, X, data, levels = 24, cmap = 'magma', zorder = 0)

ax.hlines(17, 5, 23 , "#000000", lw = 4)
ax.hlines(3, 5, 23, "#FFFFFF", lw = 4)


ax.set_xticks(np.arange(0, 29, 2))
ax.set_xticks(np.arange(1, 29, 2), minor=True)

ax.tick_params(axis='x', which='major', length=8)
ax.tick_params(axis='x', which='minor', length=4)

ax.set_yticks(np.arange(0, 21, 2))
ax.set_yticks(np.arange(1, 21, 2), minor=True)

ax.tick_params(axis='y', which='major', length=8)
ax.tick_params(axis='y', which='minor', length=4)

ax.grid(color = "#FFFFFF", zorder = 10, alpha = 0.2, ls = (0,(6,7)))

ax.set_xlabel("x (cm)")
ax.set_ylabel("y (cm)")


plt.savefig("casoDosEquipotencialesCondensador.png", dpi = 300)
plt.show()
```
