```python
import numpy as np
import scipy
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colormaps
import math
#from acrylic import Color
```

```python
dfNa = pd.read_excel('OpticaP2.xlsx', sheet_name='Na')
dfRed = pd.read_excel('OpticaP2.xlsx', sheet_name='Red')
dfBlu = pd.read_excel('OpticaP2.xlsx', sheet_name='Blu')
```

```python
L_Na = dfNa['L_D']
L0_Na = dfNa['L0_D']

L_Red = dfRed['L_D']
L0_Red = dfRed['L0_D']

L_Blu = dfBlu['L_D']
L0_Blu = dfBlu['L0_D']
```

```python
alpha = np.pi/3

deltaM_Na = np.abs(L_Na-L0_Na)*np.pi/180
deltaM_Red = np.abs(L_Red-L0_Red)*np.pi/180
deltaM_Blu = np.abs(L_Blu-L0_Blu)*np.pi/180

n_Na = np.sin((deltaM_Na+alpha)/2)/np.sin(alpha/2)
n_Red = np.sin((deltaM_Red+alpha)/2)/np.sin(alpha/2)
n_Blu = np.sin((deltaM_Blu+alpha)/2)/np.sin(alpha/2)
```
