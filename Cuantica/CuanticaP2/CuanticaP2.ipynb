```python
import numpy as np
import scipy
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker
#from matplotlib import colormaps
#import math
```

```python
df = pd.read_excel("CuanticaP2.xlsx", sheet_name="Alex")

df2 = pd.read_excel("CuanticaP2parte2.xlsx", sheet_name="Alex2")
```

```python
f = df["f"].to_numpy()
V = df["V"].to_numpy()
Verr = 0.005*V+0.003

Vyellow = df2["Vyellow"].to_numpy()
Vyellowerr = 0.005*Vyellow+0.003

Vgreen = df2["Vgreen"].to_numpy()
Vgreenerr = 0.005*Vgreen+0.003

Vblue = df2["Vblue"].to_numpy()
Vblueerr = 0.005*Vblue+0.003

Vviolet1 = df2["Vviolet1"].to_numpy()
Vviolet1err = 0.005*Vviolet1+0.003

Vviolet2 = df2["Vviolet2"].to_numpy()
Vviolet2err = 0.005*Vviolet2+0.003
```

```python
colors = ["#F1C232",
          "#6AA84F",
          "#3C78D8",
          "#3D85C6",
          "#8E7CC3"]

errorcolors = ["#B19212",
          "#4A782F",
          "#2C48B2",
          "#2565A2",
          "#6E5C93"]
labels = [""]
```

```python
def lineal(x, a, b):
    return a * x + b

popt, pcov = scipy.optimize.curve_fit(lineal, f, V, sigma = Verr)
m,n = popt
merr, nerr = np.sqrt(np.diag(pcov))

plt.plot(f, lineal(f, m, n), color = "#AA0000")
for i in range(len(f)):
    plt.errorbar(f[i], V[i], yerr = Verr[i], color = colors[i], ecolor= errorcolors[i], marker = ".", linestyle = "None", capsize = 5, zorder = 10)

plt.ylabel("V₀ (V)")
plt.xlabel("ν (Hz)")

plt.grid(True)

plt.savefig("CuanticaP2.png", dpi = 300)
```

```python
#Calculate chicuadradoreducido
residuals = (V - lineal(f, m, n))
chi2 = np.sum((residuals/Verr)**2)
chi2_red = chi2/(len(V)-len(popt))

#Calculate errecuadrado
ss_res = np.sum(residuals ** 2)
ss_tot = np.sum((V - np.mean(V)) ** 2)
r_squared = 1 - (ss_res / ss_tot)

print("Chi^2 reducido = " + str(chi2_red))
print("R^2 = " + str(r_squared))
```

```python
e = 1.60217663*10**-19

h = m*e
herr = e*merr

f0 = -n/m
f0err = np.sqrt((-nerr/m)**2+(n*merr/m**2)**2)

w0 = f0 * h
werr = nerr

hTheo = 6.62607015 * 10 ** -34
hRelativeError = abs(hTheo - h) / hTheo


print("m = " + str(m) + " pm " + str(merr))
print("n = " + str(n) + " pm " + str(nerr))
print("h = " + str(h) + " pm " + str(herr) + "J·s, abs.err = " + str(hRelativeError*100) + "%")
print("f_0 = " + str(f0/10**14) + " pm " + str(f0err/10**14) + "10^14 Hz")
print("w_0 = " + str(w0 / e) + " pm " + str(werr) + " eV")
```

```python
RT = [1,0.8,0.6,0.4,0.2]

fig, ax = plt.subplots()

#for i in range(len(f)):
 #   plt.errorbar(1, V[i], yerr = Verr[i], color = colors[i], marker = ".", linestyle = "None", capsize = 5, zorder = 10)

plt.errorbar(RT, Vyellow, yerr = Vyellowerr, color = colors[0], marker = "d", linestyle = "None", capsize = 5, zorder = 10, ecolor = errorcolors[0], barsabove=False)
plt.errorbar(RT, Vgreen, yerr = Vgreenerr, color = colors[1], marker = "o", linestyle = "None", capsize = 5, zorder = 10, ecolor = errorcolors[1], barsabove=False)
plt.errorbar(RT, Vblue, yerr = Vblueerr, color = colors[2], marker = "p", linestyle = "None", capsize = 5, zorder = 10, ecolor = errorcolors[2], barsabove=False)
plt.errorbar(RT, Vviolet1, yerr = Vviolet1err, color = colors[3], marker = "P", linestyle = "None", capsize = 5, zorder = 10, ecolor = errorcolors[3], barsabove=False)
plt.errorbar(RT, Vviolet2, yerr = Vviolet2err, color = colors[4], marker = "X", linestyle = "None", capsize = 5, zorder = 10, ecolor = errorcolors[4], barsabove=False)

plt.grid()

ax.xaxis.set_major_locator(ticker.FixedLocator([0, 0.2, 0.4, 0.6, 0.8, 1]))
ax.xaxis.set_major_formatter(ticker.PercentFormatter(xmax = 1))

plt.xlabel("Relative Transmission")
plt.ylabel("Stopping Voltage (V)")

plt.savefig("VoltageTransmission.png", dpi = 300)
```
