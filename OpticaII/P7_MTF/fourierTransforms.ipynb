```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 500
n = 50
R = 20

XX = np.linspace(0,N-1,N)
YY = np.linspace(0,N-1,N)

X, Y = np.meshgrid(XX, YY)
```

```python
def slitFunc(X, Y):
    return np.where(X % n < n/2, 1, 0)

def circleGridFunc(X,Y):
    return np.where(((X % n - n/2)**2 + (Y % n - n/2)**2 < R**2), 1, 0)
```

```python
slitArray = slitFunc(X,Y)

circleGridArray = circleGridFunc(X,Y)
```

```python
slitTransform = np.fft.fftn(slitArray)
circleTransform = np.fft.fftn(circleGridArray)
```

```python
plt.imshow(np.abs(np.fft.fftshift(slitTransform)), cmap ='inferno')

plt.show()

plt.imshow(slitArray, cmap = 'inferno')

plt.show()
```

```python
fig, ax = plt.subplots(dpi = 250)

ax.imshow(np.pow(np.abs(np.fft.fftshift(circleTransform)),0.01), cmap ='inferno')

plt.show()

plt.imshow(circleGridArray, cmap = 'inferno')

plt.show()
```
