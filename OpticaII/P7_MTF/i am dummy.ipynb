```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
resolution = 1000
N = 100
k = 1
d = 5
a = 1

XX = np.linspace(-np.pi/2,np.pi/2,resolution)
YY = np.linspace(-.1,.1,resolution)

X, Y = np.meshgrid(XX, YY)
```

```python
def transformadaRendija(X,Y):
    return (np.sin(k*N*d*X)/np.sin(k*d*X))**2 * (np.sin(k*a*X)/(k*a*X))**2
```

```python
def transformadaCirculo(X,Y):
    R2 = X**2 + Y**2
    return (np.sin(k*Nx*d*X)/np.sin(k*d*X))**2 * (np.sin(k*Ny*d*Y)/np.sin(k*d*Y))**2 * (2*scipy.special.jv(1,k*a*R2)/(k*a*R2))**2
```

```python
slitTransformArray = transformadaRendija(X,Y)
```

```python
#plt.plot(XX, slitTransformArray[0])
#plt.show()

fig, ax = plt.subplots(dpi = 250)

ax.imshow(np.sqrt(slitTransformArray), cmap = "gray", extent = (min(XX),max(XX),min(YY),max(YY)), vmin = 0, vmax = 50)

ax.set_yticklabels([])
ax.set_yticks([])

plt.savefig("fourierDiffraccion.png", bbox_inches = "tight")

plt.show()
```
